## zadanie 1

```python
import numpy as np, matplotlib.pyplot as plt
```

```python
def gen_xt(n, x0, phi, sigma):
  X = np.zeros(n)
  X[0] = x0
  for i in range(1, n):
    noise = np.random.normal(0, sigma)
    X[i] = phi * X[i - 1] + noise
  return X
x0 = 0
phi = 0.2
sigma = np.sqrt(0.4)
n = 1000
X = gen_xt(n, x0, phi, sigma)
```

```python
plt.plot(range(n), X, color = 'hotpink')
```

```python
def sampleACVF(X, h):
    n = len(X)
    x_mean = np.mean(X)
    return 1/n*sum((X[i]-x_mean)*(X[i+abs(h)]-x_mean) for i in range(1,n-abs(h)))
```

```python
def teo_acvf(phi, h):
    return phi**abs(h)*sigma**2/(1-phi**2)
```

```python
s_acvf = np.array([sampleACVF(X, h) for h in range(-50, 50)])
t_acvf = np.array([teo_acvf(phi, h) for h in range(-50, 50)])
plt.scatter(range(-50, 50), s_acvf, color = 'hotpink')
plt.scatter(range(-50, 50), t_acvf, color = 'mediumseagreen', marker = '*'); plt.grid()
```

```python
def teo_acf(phi, h):
    return phi**abs(h)
def sampleACF(X, h):
    return sampleACVF(X, h)/sampleACVF(X, 0)
```

```python
s_acf = np.array([sampleACF(X, h) for h in range(-50, 50)])
t_acf = np.array([teo_acf(phi, h) for h in range(-50, 50)])
plt.scatter(range(-50, 50), s_acf, color = 'hotpink')
plt.scatter(range(-50, 50), t_acf, color = 'mediumseagreen', marker = '*'); plt.grid()
```

## zadanie 2

```python
mc = 100
n = 1000
hs = np.arange(-20, 20)
X = [gen_xt(n, x0, phi, sigma) for i in range(mc)]
A = np.array([[sampleACF(X[i], h) for h in hs] for i in range(mc)])
q95 = np.quantile(A, 0.95, axis = 0)
q5 = np.quantile(A, 0.05, axis = 0)
plt.plot(hs, q95, color = 'hotpink', linestyle = '--')
plt.plot(hs, q5, color = 'hotpink', linestyle = '--')

```

```python
T = gen_xt(n, x0, phi, sigma)
acf = [sampleACF(T, h) for h in hs]
tacf = [teo_acf(phi, h) for h in hs]
plt.plot(hs, q95, color = 'k', linestyle = '--')
plt.plot(hs, q5, color = 'k', linestyle = '--')
plt.scatter(hs, acf, color = 'hotpink', marker = 'o')
plt.scatter(hs, tacf, color = 'mediumseagreen', marker = '*'); plt.grid()
plt.title('ACF')
```

```python
ACV = np.array([[sampleACVF(X[i], h) for h in hs] for i in range(mc)])
```

```python
vq95 = np.quantile(ACV, 0.95, axis = 0)
vq5 = np.quantile(ACV, 0.05, axis = 0)
acvf = [sampleACVF(T, h) for h in hs]
tacvf = [teo_acvf(phi, h) for h in hs]
plt.plot(hs, vq95, color = 'k', linestyle = '--')
plt.plot(hs, vq5, color = 'k', linestyle = '--')
plt.scatter(hs, acvf, color = 'hotpink', marker = 'o')
plt.scatter(hs, tacvf, color = 'mediumseagreen', marker = '*'); plt.grid()
plt.title('ACVF')
```
